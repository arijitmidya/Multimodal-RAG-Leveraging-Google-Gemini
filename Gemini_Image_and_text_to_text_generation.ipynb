```python
import google.generativeai as genai
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import Markdown
```

```python
GOOGLE_API_KEY ="add your gemini key here"
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# Define the image path
image_path = 'map.png'

# Open and display the image
try:
    image = Image.open(image_path)
    plt.imshow(image)
    # Turn off axis numbers and ticks
    plt.axis('off')
    plt.show()
except IOError:
    print("Unable to open image")
```

```python

# Print the details file
with open("details.txt", 'r') as f:
  markdown_text = f.read()
output(markdown_text)
```

```python
# Upload the text file
details_file = genai.upload_file(path="details.txt", display_name="details", mime_type="text/markdown")
```

```python
# Use gemini-1.5-flash to generate a description

model = genai.GenerativeModel('gemini-1.5-flash')
response = model.generate_content(["Please analyze the provided map image and text file",details_file, image])
```

```python
def output(text):
  text = text.replace('•', '  *')
  return Markdown(text)

output(response.text)
```
